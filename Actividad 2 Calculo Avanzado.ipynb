# Actividad N°2: Derivada Direccional.


 ¿Cuál es la segunda derivada direccional (1) para la forma cuadrática?
 
 Se tiene por definicion que la derivada direccional está dada por la fórmula: 
 
 $$ D _{u} f(x)= {u} \cdot \nabla f(x) $$
 
 En donde:
 
 $ {u} $ : Corresponde al vector unitario. 
 $ \newline $
 $ \nabla f(x) $ : Corresponde al vector gradiente. 
 
 Además se sabe que la forma cuadrática está dada por: 
 
 $$ f({x_1} , {x_2}) = \frac{1}{2}{x^T}{A}{x} + {x^T}{b}+c $$

 
 En donde :
 
 $ {A} $ :  Es una matriz simétrica cuadrada. 
 $ \newline $
 $ A =  \begin{pmatrix} a_{11} & a_{12} \\ a_{12} & a_{22} \end{pmatrix}$
 $ \newline $
 $ {x} $ : Es un vector. 
 $ \newline\ $
 $ x = \begin{pmatrix} x_{1} \\ x_{2} \end{pmatrix}$
 $ \newline\ $
 $ {b} $ : Es un vector. 
 $ \newline\ $
 $ b = \begin{pmatrix} b_{1} \\ b_{2} \end{pmatrix}$
 $ \newline\ $
 $ {c} $ : Es una constante.
 
Para la obtención de la segunda derivada usaremos las definiciones anteriores.
Primero calculamos el vector gradiente por medio de las derivadas parciales con respecto a $ x_{1} , x_{2} $


(1)
$\frac{\partial{u}}{\partial x_{1}} f(x_{1})= \frac{\partial{u}}{\partial x_{1}} (\frac{a_{11}x^2_{1}}{2} + {a_{12}}{x_{1}}{x_{2}} + \frac{a_{22}x^2_{2}}{2} +{b_{1}}{x_{1}} +{b_{2}}{x_{2}} + {c}) $

Resolviendo se obtiene que la derivada parcial con respecto a $ x_{1} $ = $\frac{\partial{u}}{\partial x_{1}} f(x_{1}) = {a_{11}}{x_{1}}+{a_{12}} {x_{2}} + {b_{1}} $ 

Ahora, la derivada parcial con respecto a $x_{2} $ está dada por la primera fórmula (1) pero esta vez evaluamos con respecto a $x_{2} $ quedándonos:

$\frac{\partial{u}}{\partial x_{2}} f(x_{2})= {a_{12}}{x_{1}}+{a_{22}} {x_{2}} + {b_{2}} $

Ordenando las componentes:

$ \nabla f(x) = \begin{pmatrix} f_{x_{1}}({x_{1},x_{2}})  \\ f_{x_{2}}({x_{1},x_{2}})  \end{pmatrix} = \begin{pmatrix} a_{11} & a_{12} \\ a_{12} & a_{22} \end{pmatrix}\begin{pmatrix} x_{1} \\ x_{2} \end{pmatrix} +\begin{pmatrix} b_{1} \\ b_{2} \end{pmatrix} = {A}{x}+{b}$ 

Por lo tanto el vector gradiente está dado por $
\begin{equation}
\nabla f(x) ={A}{x}+{b} \newline\
\end{equation}$
Aplicando en la ecuación : $D _{u} f(x)= {u} \cdot \nabla f(x) = \newline\ $
$D _{u} f(x)= {u} \cdot {A}{x}+{b}. $ Como se mencionó anteriormente, ${u}$ corresponde al vector unitario $({u_{1}},{u_{2}})$

De esta forma encontramos la primera derivada direccional para la forma cuadrática. Para obtener la segunda, repetiremos los  pasos anteriores pero esta vez aplicando a la primera derivada direccional obtenida.

Aplicaremos derivadas parciales con respecto a  $ {x_{1}}, {x_{2}} $ :
$
\begin{equation}
\frac{\partial{u}}{\partial x_{1}} f(x_{1})= \frac{\partial{u}}{\partial x_{1}}({a_{11}}{x_{1}}+{a_{12}}{x_{2}}+{b_{1}}+{a_{12}}{x_{1}}+{a_{22}}{x_{2}}+{b_{2}})\newline\ 
\end{equation}$
Resolviendo se obtiene que la derivada parcial con respecto a $ x_{1} $ = $
\begin{equation}
\frac{\partial{u}}{\partial x_{1}} f(x_{1}) = {a_{11}}+{a_{12}}
\end{equation}$

Con respecto a $ 
\begin{equation}
{x_{2}} = \frac{\partial{u}}{\partial x_{2}}f(x_{2})=\frac{\partial{u}}{\partial x_{2}} ({a_{11}}{x_{1}}+{a_{12}}{x_{2}}+{b_{1}}+{a_{12}}{x_{1}}+{a_{22}}{x_{2}}+{b_{2}}) \newline\ 
\end{equation}$
Resolviendo se obtiene que la derivada parcial con respecto a $ x_{2}$ = $
\begin{equation}
\frac{\partial{u}}{\partial x_{1}} f(x_{2})= {a_{12}}+{a_{22}}
\end{equation}$

Ordenando las componentes nos queda la matriz:

$\begin{pmatrix} a_{11} & a_{12} \\ a_{12} & a_{22} \end{pmatrix} $ La cual está multiplicada por el vector unitario $ {u} $

Expresada de otra forma : $ {A} \cdot {u} $ .
Así obtenemos la segunda derivada direccional de la forma cuadrática la cual queda expresada como:

$ D _{u} f(x)= {u} \cdot {A} $ En donde $ {u} $ corresponde al vector unitario $({u_{1}},{u_{2}})$ .
















 
 

 

Se define la segunda derivada de la función f(x,y) en la dirección a (ángulo formado con el eje X), con vector unitario u\alpha, como:
\begin{equation*}
    D_{\alpha}^{2}f(x,y)=u_{\alpha}^{T}*H*u_{\alpha}
\end{equation*}


siendo H la matriz Hessiana
\begin{bmatrix}
\frac{\partial ^2f(x,y)}{\partial x^2} & \frac{\partial ^2f(x,y)}{\partial x\partial y} \\ 
\frac{\partial ^2f(x,y)}{\partial x\partial y} & \frac{\partial ^2f(x,y)}{\partial y^2}
\end{bmatrix}

Si se toma como ua el vector unitario en la dirección del gradiente uq, se obtiene
\begin{equation}
D_{0}^{2}f(x,y)=\frac{\partial^2 f(x,y)}{\partial x^2}cos^2\theta +2\frac{\partial ^2f(x,y)}{\partial x\partial y}sen\theta cos\theta +\frac{\partial ^2f(x,y)}{\partial y^2}sen^2\theta 
\end{equation}

o también su discretización
\begin{equation}
    D_{0}^{2}=\frac{\Delta _{xx}f*(\Delta _{x}f)^2+2*\Delta _{xy}f*\Delta _{x}f*\Delta _{y}f+\Delta _{yy}f*(\Delta_{y}f)^2 }{(\Delta_{x}f)^2+(\Delta _{y}f)^2 }
\end{equation}
